```python
set_parameter = {
    'sep': ',',
    'encoding': 'ISO-8859-1'
}
analizer = Analizer("data/merge_train.csv", **set_parameter)
analizer.head()
```

```python
MAX_NULL_FOR_ROW = 3
data = analizer.data.copy()
data.shape
```

```python
"""
Lo que se realiza a continuacion es eliminar todas las filas que contengan
un numero de variables nulas superior a MAX_NULL_FOR_ROW
recordar:
sum(axis=1) indica que sumara los valores por columna
data.isnull() formara un dataframe indicando si cada elemnto presente es
              nulo o no (False, True)
"""
data_cut = data.loc[data.isnull().sum(axis=1) <= MAX_NULL_FOR_ROW]
data_cut.shape
```

```python
analizer.data = data_cut
analizer.null_verificator()
```
