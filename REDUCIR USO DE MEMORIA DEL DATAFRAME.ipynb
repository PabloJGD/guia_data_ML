```python
#importaciones
import pandas as pd

#variables 
set_parameter = {
    'sep': ',',
    'encoding': 'ISO-8859-1',
    'low_memory': False
}

data = pd.read_csv('data/merge_train.csv', **set_parameter)
data.head(5)
```

```python
def get_memory_usage(data):
    return '{} MB'.format(data.memory_usage(deep=True).sum() / 1024 ** 2)

get_memory_usage(data)
```

```python
import numpy as np
def reduce_size_data(df, save=True, new_name='data_qroma_train_reduce.csv'):
    for col in df.select_dtypes(include=['int']).columns:
        df[col] = df[col].apply(pd.to_numeric, downcast='integer')
 
    for col in df.select_dtypes(include=['float']).columns:
        df[col] = df[col].apply(pd.to_numeric, downcast='float')

    for col in df.select_dtypes(include=['object']).columns:
        if len(df[col].unique()) / len(df[col]) < 0.5:
            df[col] = df[col].astype('category')
    
reduce_size_data(data)
get_memory_usage(data)
```
